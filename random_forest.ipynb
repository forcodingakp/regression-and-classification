```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Random_forest\images_analyzed_productivity1.csv")
```

```python
df.head()
```

```python
sizes = df['Productivity'].value_counts(sort=1)
print(sizes)
```

```python
print(df.columns)
```

```python
df = df.drop(['Images_Analyzed'], axis=1, errors='ignore')
```

```python
df.head()
```

```python
df.drop(['User'], axis=1, inplace=True)
```

```python
#conversion of non numeric data to numeric
df.Productivity[df.Productivity == 'Good'] = 1
df.Productivity[df.Productivity == 'Bad'] = 2

df.head()
```

```python
Y = df['Productivity'].values

Y = Y.astype('int')
```

```python
X = df.drop(labels=['Productivity'], axis=1)
```

```python
%pip install scikit-learn
```

```python

from sklearn.model_selection import train_test_split

X_train,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=20)
```

```python
print(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10, random_state=30)
model.fit(X_train, Y_train)
```

```python
prediction_test = model.predict(X_test)
print(prediction_test)
```

```python
from sklearn import metrics
print("Accuracy = ", metrics.accuracy_score(Y_test, prediction_test))
```

```python
feature_list = list(X.columns)
```

```python
feature_imp = pd.Series(model.feature_importances_, index=feature_list).sort_values(ascending=False)
```
